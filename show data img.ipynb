```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
```

```python
img_path = './data/val/original/*'
```

```python
files = glob.glob(img_path)
```

```python
plt.figure(figsize=(30,8))
w = 4
n = 40
h = int(n / w)
for i,d in enumerate(files[:n]):
    plt.subplot(h,w,i+1)
    plt.imshow(mpimg.imread(d))
```
